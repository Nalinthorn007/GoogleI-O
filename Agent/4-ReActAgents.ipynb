#### ReAct Architecture ຄືຍັງ ?

 <strong>ການເຮັດວຽກ ReAct</strong>
 - ການຄິດ (Thought): ວິເຄາະສະຖານະການປັດຈຸບັນ ແລະ ວາງແຜນຂັ້ນຕອນຕໍ່ໄປ
 - ການກະທຳ (Action): ໃຊ້ເຄື່ອງມືຕ່າງໆ ເຊັ່ນ ຄົ້ນຫາ, ຄຳນວນ, ຫຼື ເຂົ້າເຖິງຂໍ້ມູນ
 - ການສັງເກດ (Observation): ຮັບຜົນຈາກການກະທຳ ແລະ ປັບແຜນຕາມຄວາມຈຳເປັນ

#### ອົງປະກອບ 4 ຫຼັກຂອງ LangGraph
 <img src="./Image/MemorySaver.png" alt="MemorySaver" width="1400">

1. Graph
 - ມັນຄື: ແຜນວຽກໂດຍລວມ ກຳນົດໂຄງສ້າງຂອງຂໍ້ມູນທີ່ໄຫຼຜ່ານ Graph

  <strong> ໜ້າທີ່: </strong>

 - ກຳນົດວ່າ Node ໃດມາກ່ອນ-ຫຼັງ
 - ຄວບຄຸມການໄຫຼຂອງຂໍ້ມູນ 

2. Super-steps
 - ມັນຄື: ກຸ່ມຂອງການເຮັດວຽກທີ່ເກີດຂຶ້ນພ້ອມກັນ ໃນແຕ່ລະໂນ້ດ

  <strong> ໜ້າທີ່: </strong> 

 Super-steps (ຂັ້ນຕອນການເຮັດວຽກ)

 1. Super-step 1 (Sequential - ຕິດຕໍ່ກັນ):
   - Node A: ກວດສອບເມນູ
   - ຖ້າສຳເລັດ → ໄປ Super-step 2

 2. Super-step 2 (Parallel - ພ້ອມກັນ):
  - Node B: ກວດສະຕ໋ອກແປ້ງ
  - Node C: ກວດສະຕ໋ອກເຄື່ອງປຸງ
  - Node D: ກວດສະຕ໋ອກຜັກ
  - ທຸກອັນເສັດພ້ອມກັນ → ໄປ Super-step 3

 3. Super-step 3 (Sequential):
  - Node E: ປຸງແຕ່ງອາຫານ
  - ຖ້າແລ້ວ → ໄປ Super-step 4

 4. Super-step 4 (Sequential):
  - Node F: ຈັດສົ່ງ

3. Checkpoints (ຈຸດບັນທຶກ)
 - Checkpoint 1: "ເມນູ: ຜັດໄທ, ລູກຄ້າ: ທ້າວ ນາລິນທອນ"
 - Checkpoint 2: "ເສັ້ນ: ມີພໍ, ເຄື່ອງປຸງ: ມີພໍ"
 - Checkpoint 3: "ອາຫານສຳເລັດ: ຜັດໄທ 1 ຈານ"
 - Checkpoint 4: "ຈັດສົ່ງສຳເລັດ: 14:30"

4. Thread (ປື້ມປະຫວັດ)
 - ເກັບທຸກ checkpoint ຂອງການສັ່ງອາຫານນີ້ໄວ້ - ຕັ້ງແຕ່ເລີ່ມສັ່ງຈົນເຖິງໄດ້ຮັບ




```python
from typing import TypedDict , List, Annotated
from IPython.display import Image, display 
from langgraph.graph import StateGraph, START, END
from langgraph.prebuilt import ToolNode , tools_condition 
from langgraph.checkpoint.memory import MemorySaver
import operator
```

```python
# ກຳນົດ State structure ດ້ວຍ Annotated keys
class OrderState(TypedDict):
    order: str
    ingredients_checked: bool
    food_cooked: bool
    delivered: bool
    status: str
    # ເພີ່ມການຕິດຕາມແຕ່ລະວັດຖຸດິບ
    rice_available: bool
    vegetables_available: bool
    spices_available: bool
    # ໃຊ້ Annotated ເພື່ອຮັບຫຼາຍ values
    messages: Annotated[List[str], operator.add]

# Node 1: ກວດສອບເມນູ (Sequential)
def check_menu(state: OrderState) -> OrderState:
    print(f"🍽️ Super-step 1: ກວດສອບເມນູ - {state['order']}")
    return {
        "status": "ເມນູຢືນຢັນແລ້ວ",
        "messages": [f"ເມນູ {state['order']} ຢືນຢັນແລ້ວ"]
    }

# Node 2, 3, 4: ກວດສະຕ໋ອກ (Parallel - Super-step 2)
def check_rice(state: OrderState) -> OrderState:
    rice_available = True  
    
    if rice_available:
        print("🍚 ກວດສະຕ໋ອກເຂົ້າ: ມີພໍ")
        return {
            "rice_available": True,
            "messages": ["✅ ເຂົ້າມີພໍໃຊ້"]
        }
    else:
        print("🍚 ກວດສະຕ໋ອກເຂົ້າ: ໝົດແລ້ວ!")
        return {
            "rice_available": False,
            "messages": ["❌ ເຂົ້າໝົດແລ້ວ - ບໍ່ສາມາດປຸງໄດ້"]
        }

def check_vegetables(state: OrderState) -> OrderState:
    vegetables_available = True  
    
    if vegetables_available:
        print("🥬 ກວດສະຕ໋ອກຜັກ: ມີພໍ")
        return {
            "vegetables_available": True,
            "messages": ["✅ ຜັກມີພໍໃຊ້"]
        }
    else:
        print("🥬 ກວດສະຕ໋ອກຜັກ: ໝົດແລ້ວ!")
        return {
            "vegetables_available": False,
            "messages": ["❌ ຜັກໝົດແລ້ວ - ບໍ່ສາມາດປຸງໄດ້"]
        }

def check_spices(state: OrderState) -> OrderState:
    spices_available = False 
    
    if spices_available:
        print("🧂 ກວດເຄື່ອງປຸງ: ຄົບຖ້ວນ")
        return {
            "spices_available": True,
            "messages": ["✅ ເຄື່ອງປຸງຄົບຖ້ວນ"]
        }
    else:
        print("🧂 ກວດເຄື່ອງປຸງ: ຂາດແລ້ວ!")
        return {
            "spices_available": False,
            "messages": ["❌ ເຄື່ອງປຸງບໍ່ຄົບ - ບໍ່ສາມາດປຸງໄດ້"]
        }

# Node 5: ປຸງແຕ່ງ (Sequential - Super-step 3)
def cook_food(state: OrderState) -> OrderState:
    print(f"👨‍🍳 Super-step 3: ກວດສອບວັດຖຸດິບສຳລັບ {state['order']}...")
    print(f"📝 ສະຖານະຂອງວັດຖຸດິບ: {', '.join(state.get('messages', []))}")
    
    # ກວດວ່າວັດຖຸດິບຄົບບໍ່
    rice_ok = state.get('rice_available', False)
    vegetables_ok = state.get('vegetables_available', False) 
    spices_ok = state.get('spices_available', False)
    
    if rice_ok and vegetables_ok and spices_ok:
        print(f"✅ ວັດຖຸດິບຄົບຖ້ວນ - ເລີ່ມປຸງ {state['order']}...")
        return {
            "ingredients_checked": True,
            "food_cooked": True,
            "status": "ອາຫານສຳເລັດ",
            "messages": [f"✅ {state['order']} ປຸງສຳເລັດແລ້ວ"]
        }
    else:
        missing_items = []
        if not rice_ok: missing_items.append("ເຂົ້າ")
        if not vegetables_ok: missing_items.append("ຜັກ")
        if not spices_ok: missing_items.append("ເຄື່ອງປຸງ")
        
        print(f"❌ ວັດຖຸດິບບໍ່ຄົບ: ຂາດ {', '.join(missing_items)}")
        print("🛑 ບໍ່ສາມາດປຸງອາຫານໄດ້!")
        
        return {
            "ingredients_checked": False,
            "food_cooked": False,
            "status": f"ບໍ່ສາມາດປຸງໄດ້ - ຂາດ {', '.join(missing_items)}",
            "messages": [f"❌ {state['order']} ປຸງບໍ່ໄດ້ - ຂາດວັດຖຸດິບ"]
        }

# Node 6: ຈັດສົ່ງ (Sequential - Super-step 4)
def deliver(state: OrderState) -> OrderState:
    if state.get('food_cooked', False):
        print(f"🚗 Super-step 4: ຈັດສົ່ງ {state['order']} ສຳເລັດ")
        return {
            "delivered": True,
            "status": "ຈັດສົ່ງແລ້ວ",
            "messages": [f"🚗 {state['order']} ຈັດສົ່ງສຳເລັດ"]
        }
    else:
        print(f"❌ ບໍ່ສາມາດຈັດສົ່ງໄດ້ - ອາຫານຍັງບໍ່ສຳເລັດ")
        return {
            "delivered": False,
            "status": "ຈັດສົ່ງບໍ່ໄດ້ - ອາຫານບໍ່ສຳເລັດ",
            "messages": [f"❌ {state['order']} ຈັດສົ່ງບໍ່ໄດ້"]
        }

# ສ້າງ Graph
def create_order_graph():
    # ໃຊ້ MemorySaver ເປັນ checkpointer
    memory = MemorySaver()
    
    # ສ້າງ StateGraph
    workflow = StateGraph(OrderState)
    
    # ເພີ່ມ Nodes
    workflow.add_node("check_menu", check_menu)
    workflow.add_node("check_rice", check_rice)
    workflow.add_node("check_vegetables", check_vegetables) 
    workflow.add_node("check_spices", check_spices)
    workflow.add_node("cook_food", cook_food)
    workflow.add_node("deliver", deliver)
    
    # ກຳນົດ Edges (ການເຊື່ອມຕໍ່)
    # Super-step 1 → Super-step 2
    workflow.set_entry_point("check_menu")
    workflow.add_edge("check_menu", "check_rice")
    workflow.add_edge("check_menu", "check_vegetables")
    workflow.add_edge("check_menu", "check_spices")
    
    # Super-step 2 → Super-step 3
    workflow.add_edge("check_rice", "cook_food")
    workflow.add_edge("check_vegetables", "cook_food")
    workflow.add_edge("check_spices", "cook_food")
    
    # Super-step 3 → Super-step 4
    workflow.add_edge("cook_food", "deliver")
    workflow.add_edge("deliver", END)
    
    # Compile ດ້ວຍ checkpointer
    app = workflow.compile(checkpointer=memory)
    display(Image(app.get_graph().draw_mermaid_png()))
    return app

# ການນຳໃຊ້
def main():
    # ສ້າງ Graph
    app = create_order_graph()
    
    # ສ້າງ thread_id ສຳລັບ session
    thread_config = {"configurable": {"thread_id": "order_124"}}
    
    # State ເບື້ອງຕົ້ນ
    initial_state = {
        "order": "ຜັດໄທ",
        "ingredients_checked": False,
        "food_cooked": False,
        "delivered": False,
        "status": "ເລີ່ມສັ່ງອາຫານ",
        "rice_available": False, 
        "vegetables_available": False, 
        "spices_available": False,
        "messages": []
    }
    
    print("=" * 50)
    print("🚀 ເລີ່ມການສັ່ງອາຫານ")
    print("=" * 50)
    
    # ເຮັດວຽກ
    result = app.invoke(initial_state, config=thread_config)
    
    print("\n" + "=" * 50)
    print("📋 ສະຖານະສຸດທ້າຍ:")
    print(f"   ອາຫານ: {result['order']}")
    print(f"   ສະຖານະ: {result['status']}")
    print(f"   ວັດຖຸດິບ - ເຂົ້າ: {'✅' if result.get('rice_available') else '❌'}")
    print(f"   ວັດຖຸດິບ - ຜັກ: {'✅' if result.get('vegetables_available') else '❌'}")
    print(f"   ວັດຖຸດິບ - ເຄື່ອງປຸງ: {'✅' if result.get('spices_available') else '❌'}")
    print(f"   ກວດສະຕ໋ອກ: {result['ingredients_checked']}")
    print(f"   ປຸງແລ້ວ: {result['food_cooked']}")
    print(f"   ສົ່ງແລ້ວ: {result['delivered']}")
    print(f"   ຂໍ້ຄວາມທັງໝົດ: {result['messages']}")
    

if __name__ == "__main__":
    main()
```

# ຕົວຢ່າງ

```python
from langchain_community.tools import ArxivQueryRun,WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper,ArxivAPIWrapper
from langchain_community.tools.tavily_search import TavilySearchResults
from dotenv import load_dotenv
from langchain_core.messages import AnyMessage 
from langgraph.graph.message import add_messages
from langgraph.checkpoint.memory import MemorySaver
from langchain_core.messages import HumanMessage
load_dotenv()

import os

os.environ["TAVILY_API_KEY"]=os.getenv("TAVILY_API_KEY")
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY") 
```

```python
class State(TypedDict):
    messages:Annotated[list[AnyMessage],add_messages]
```

```python
api_wrapper_arxiv=ArxivAPIWrapper(top_k_results=2,doc_content_chars_max=500)
arxiv=ArxivQueryRun(api_wrapper=api_wrapper_arxiv) 
arxiv.invoke("Attention iss all you need")
```

```python
api_wrapper_wiki=WikipediaAPIWrapper(top_k_results=1,doc_content_chars_max=500)
wiki=WikipediaQueryRun(api_wrapper=api_wrapper_wiki)
wiki.invoke("What is machine learning")
```

```python
tavily = TavilySearchResults()
tavily.invoke("Provide me the recent AI news for july 27th 2025")
```

```python
### Custom Functions
def multiply(a: int, b: int) -> int:
    """Multiply a and b.

    Args:
        a: first int
        b: second int
    """
    return a * b

# This will be a tool
def add(a: int, b: int) -> int:
    """Adds a and b.

    Args:
        a: first int
        b: second int
    """
    return a + b

def divide(a: int, b: int) -> float:
    """Divide a and b.

    Args:
        a: first int
        b: second int
    """
    return a / b
```

```python
#set Tools
tools=[arxiv,wiki,tavily,add,divide,multiply]
```

```python
from langchain_groq import ChatGroq
llm=ChatGroq(model="qwen/qwen3-32b")
llm_with_tools = llm.bind_tools(tools) 
```

```python
### Node definition
def tool_calling_llm(state:State):
    return {"messages":[llm_with_tools.invoke(state["messages"])]}

# Build graph
builder = StateGraph(State)
builder.add_node("tool_calling_llm", tool_calling_llm)
builder.add_node("tools", ToolNode(tools))

builder.add_edge(START, "tool_calling_llm")
builder.add_conditional_edges(
    "tool_calling_llm",
    # If the latest message (result) from assistant is a tool call -> tools_condition routes to tools
    # If the latest message (result) from assistant is a not a tool call -> tools_condition routes to END
    tools_condition,
)
builder.add_edge("tools","tool_calling_llm") 
```

```python

memory=MemorySaver()
graph_memory = builder.compile(checkpointer=memory)

# View
display(Image(graph_memory.get_graph().draw_mermaid_png()))
```

```python
## Specify the thread

config={"configurable":{"thread_id":"2"}}
# Specify an input
messages = [HumanMessage(content="Add 12 and 13.")]
messages=graph_memory.invoke({"messages":messages},config=config)
for m in messages['messages']:
    m.pretty_print()
```

```python
messages = [HumanMessage(content="then multiplty that number by 2 and also summarize Google I/O 2025")]
messages=graph_memory.invoke({"messages":messages},config=config)
for m in messages['messages']:
    m.pretty_print()
```
