```python
# ‡∫Å‡∫≥‡∫ô‡∫ª‡∫î State structure ‡∫î‡ªâ‡∫ß‡∫ç Annotated keys
class OrderState(TypedDict):
    order: str
    ingredients_checked: bool
    food_cooked: bool
    delivered: bool
    status: str
    # ‡ªÄ‡∫û‡∫µ‡ªà‡∫°‡∫Å‡∫≤‡∫ô‡∫ï‡∫¥‡∫î‡∫ï‡∫≤‡∫°‡ªÅ‡∫ï‡ªà‡∫•‡∫∞‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö
    rice_available: bool
    vegetables_available: bool
    spices_available: bool
    # ‡ªÉ‡∫ä‡ªâ Annotated ‡ªÄ‡∫û‡∫∑‡ªà‡∫≠‡∫Æ‡∫±‡∫ö‡∫´‡∫º‡∫≤‡∫ç values
    messages: Annotated[List[str], operator.add]

# Node 1: ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡ªÄ‡∫°‡∫ô‡∫π (Sequential)
def check_menu(state: OrderState) -> OrderState:
    print(f"üçΩÔ∏è Super-step 1: ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡ªÄ‡∫°‡∫ô‡∫π - {state['order']}")
    return {
        "status": "‡ªÄ‡∫°‡∫ô‡∫π‡∫¢‡∫∑‡∫ô‡∫¢‡∫±‡∫ô‡ªÅ‡∫•‡ªâ‡∫ß",
        "messages": [f"‡ªÄ‡∫°‡∫ô‡∫π {state['order']} ‡∫¢‡∫∑‡∫ô‡∫¢‡∫±‡∫ô‡ªÅ‡∫•‡ªâ‡∫ß"]
    }

# Node 2, 3, 4: ‡∫Å‡∫ß‡∫î‡∫™‡∫∞‡∫ï‡ªã‡∫≠‡∫Å (Parallel - Super-step 2)
def check_rice(state: OrderState) -> OrderState:
    rice_available = True  
    
    if rice_available:
        print("üçö ‡∫Å‡∫ß‡∫î‡∫™‡∫∞‡∫ï‡ªã‡∫≠‡∫Å‡ªÄ‡∫Ç‡∫ª‡ªâ‡∫≤: ‡∫°‡∫µ‡∫û‡ªç")
        return {
            "rice_available": True,
            "messages": ["‚úÖ ‡ªÄ‡∫Ç‡∫ª‡ªâ‡∫≤‡∫°‡∫µ‡∫û‡ªç‡ªÉ‡∫ä‡ªâ"]
        }
    else:
        print("üçö ‡∫Å‡∫ß‡∫î‡∫™‡∫∞‡∫ï‡ªã‡∫≠‡∫Å‡ªÄ‡∫Ç‡∫ª‡ªâ‡∫≤: ‡ªù‡∫ª‡∫î‡ªÅ‡∫•‡ªâ‡∫ß!")
        return {
            "rice_available": False,
            "messages": ["‚ùå ‡ªÄ‡∫Ç‡∫ª‡ªâ‡∫≤‡ªù‡∫ª‡∫î‡ªÅ‡∫•‡ªâ‡∫ß - ‡∫ö‡ªç‡ªà‡∫™‡∫≤‡∫°‡∫≤‡∫î‡∫õ‡∫∏‡∫á‡ªÑ‡∫î‡ªâ"]
        }

def check_vegetables(state: OrderState) -> OrderState:
    vegetables_available = True  
    
    if vegetables_available:
        print("ü•¨ ‡∫Å‡∫ß‡∫î‡∫™‡∫∞‡∫ï‡ªã‡∫≠‡∫Å‡∫ú‡∫±‡∫Å: ‡∫°‡∫µ‡∫û‡ªç")
        return {
            "vegetables_available": True,
            "messages": ["‚úÖ ‡∫ú‡∫±‡∫Å‡∫°‡∫µ‡∫û‡ªç‡ªÉ‡∫ä‡ªâ"]
        }
    else:
        print("ü•¨ ‡∫Å‡∫ß‡∫î‡∫™‡∫∞‡∫ï‡ªã‡∫≠‡∫Å‡∫ú‡∫±‡∫Å: ‡ªù‡∫ª‡∫î‡ªÅ‡∫•‡ªâ‡∫ß!")
        return {
            "vegetables_available": False,
            "messages": ["‚ùå ‡∫ú‡∫±‡∫Å‡ªù‡∫ª‡∫î‡ªÅ‡∫•‡ªâ‡∫ß - ‡∫ö‡ªç‡ªà‡∫™‡∫≤‡∫°‡∫≤‡∫î‡∫õ‡∫∏‡∫á‡ªÑ‡∫î‡ªâ"]
        }

def check_spices(state: OrderState) -> OrderState:
    spices_available = False 
    
    if spices_available:
        print("üßÇ ‡∫Å‡∫ß‡∫î‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫õ‡∫∏‡∫á: ‡∫Ñ‡∫ª‡∫ö‡∫ñ‡ªâ‡∫ß‡∫ô")
        return {
            "spices_available": True,
            "messages": ["‚úÖ ‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫õ‡∫∏‡∫á‡∫Ñ‡∫ª‡∫ö‡∫ñ‡ªâ‡∫ß‡∫ô"]
        }
    else:
        print("üßÇ ‡∫Å‡∫ß‡∫î‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫õ‡∫∏‡∫á: ‡∫Ç‡∫≤‡∫î‡ªÅ‡∫•‡ªâ‡∫ß!")
        return {
            "spices_available": False,
            "messages": ["‚ùå ‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫õ‡∫∏‡∫á‡∫ö‡ªç‡ªà‡∫Ñ‡∫ª‡∫ö - ‡∫ö‡ªç‡ªà‡∫™‡∫≤‡∫°‡∫≤‡∫î‡∫õ‡∫∏‡∫á‡ªÑ‡∫î‡ªâ"]
        }

# Node 5: ‡∫õ‡∫∏‡∫á‡ªÅ‡∫ï‡ªà‡∫á (Sequential - Super-step 3)
def cook_food(state: OrderState) -> OrderState:
    print(f"üë®‚Äçüç≥ Super-step 3: ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö‡∫™‡∫≥‡∫•‡∫±‡∫ö {state['order']}...")
    print(f"üìù ‡∫™‡∫∞‡∫ñ‡∫≤‡∫ô‡∫∞‡∫Ç‡∫≠‡∫á‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö: {', '.join(state.get('messages', []))}")
    
    # ‡∫Å‡∫ß‡∫î‡∫ß‡ªà‡∫≤‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö‡∫Ñ‡∫ª‡∫ö‡∫ö‡ªç‡ªà
    rice_ok = state.get('rice_available', False)
    vegetables_ok = state.get('vegetables_available', False) 
    spices_ok = state.get('spices_available', False)
    
    if rice_ok and vegetables_ok and spices_ok:
        print(f"‚úÖ ‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö‡∫Ñ‡∫ª‡∫ö‡∫ñ‡ªâ‡∫ß‡∫ô - ‡ªÄ‡∫•‡∫µ‡ªà‡∫°‡∫õ‡∫∏‡∫á {state['order']}...")
        return {
            "ingredients_checked": True,
            "food_cooked": True,
            "status": "‡∫≠‡∫≤‡∫´‡∫≤‡∫ô‡∫™‡∫≥‡ªÄ‡∫•‡∫±‡∫î",
            "messages": [f"‚úÖ {state['order']} ‡∫õ‡∫∏‡∫á‡∫™‡∫≥‡ªÄ‡∫•‡∫±‡∫î‡ªÅ‡∫•‡ªâ‡∫ß"]
        }
    else:
        missing_items = []
        if not rice_ok: missing_items.append("‡ªÄ‡∫Ç‡∫ª‡ªâ‡∫≤")
        if not vegetables_ok: missing_items.append("‡∫ú‡∫±‡∫Å")
        if not spices_ok: missing_items.append("‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫õ‡∫∏‡∫á")
        
        print(f"‚ùå ‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö‡∫ö‡ªç‡ªà‡∫Ñ‡∫ª‡∫ö: ‡∫Ç‡∫≤‡∫î {', '.join(missing_items)}")
        print("üõë ‡∫ö‡ªç‡ªà‡∫™‡∫≤‡∫°‡∫≤‡∫î‡∫õ‡∫∏‡∫á‡∫≠‡∫≤‡∫´‡∫≤‡∫ô‡ªÑ‡∫î‡ªâ!")
        
        return {
            "ingredients_checked": False,
            "food_cooked": False,
            "status": f"‡∫ö‡ªç‡ªà‡∫™‡∫≤‡∫°‡∫≤‡∫î‡∫õ‡∫∏‡∫á‡ªÑ‡∫î‡ªâ - ‡∫Ç‡∫≤‡∫î {', '.join(missing_items)}",
            "messages": [f"‚ùå {state['order']} ‡∫õ‡∫∏‡∫á‡∫ö‡ªç‡ªà‡ªÑ‡∫î‡ªâ - ‡∫Ç‡∫≤‡∫î‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö"]
        }

# Node 6: ‡∫à‡∫±‡∫î‡∫™‡∫ª‡ªà‡∫á (Sequential - Super-step 4)
def deliver(state: OrderState) -> OrderState:
    if state.get('food_cooked', False):
        print(f"üöó Super-step 4: ‡∫à‡∫±‡∫î‡∫™‡∫ª‡ªà‡∫á {state['order']} ‡∫™‡∫≥‡ªÄ‡∫•‡∫±‡∫î")
        return {
            "delivered": True,
            "status": "‡∫à‡∫±‡∫î‡∫™‡∫ª‡ªà‡∫á‡ªÅ‡∫•‡ªâ‡∫ß",
            "messages": [f"üöó {state['order']} ‡∫à‡∫±‡∫î‡∫™‡∫ª‡ªà‡∫á‡∫™‡∫≥‡ªÄ‡∫•‡∫±‡∫î"]
        }
    else:
        print(f"‚ùå ‡∫ö‡ªç‡ªà‡∫™‡∫≤‡∫°‡∫≤‡∫î‡∫à‡∫±‡∫î‡∫™‡∫ª‡ªà‡∫á‡ªÑ‡∫î‡ªâ - ‡∫≠‡∫≤‡∫´‡∫≤‡∫ô‡∫ç‡∫±‡∫á‡∫ö‡ªç‡ªà‡∫™‡∫≥‡ªÄ‡∫•‡∫±‡∫î")
        return {
            "delivered": False,
            "status": "‡∫à‡∫±‡∫î‡∫™‡∫ª‡ªà‡∫á‡∫ö‡ªç‡ªà‡ªÑ‡∫î‡ªâ - ‡∫≠‡∫≤‡∫´‡∫≤‡∫ô‡∫ö‡ªç‡ªà‡∫™‡∫≥‡ªÄ‡∫•‡∫±‡∫î",
            "messages": [f"‚ùå {state['order']} ‡∫à‡∫±‡∫î‡∫™‡∫ª‡ªà‡∫á‡∫ö‡ªç‡ªà‡ªÑ‡∫î‡ªâ"]
        }

# ‡∫™‡ªâ‡∫≤‡∫á Graph
def create_order_graph():
    # ‡ªÉ‡∫ä‡ªâ MemorySaver ‡ªÄ‡∫õ‡∫±‡∫ô checkpointer
    memory = MemorySaver()
    
    # ‡∫™‡ªâ‡∫≤‡∫á StateGraph
    workflow = StateGraph(OrderState)
    
    # ‡ªÄ‡∫û‡∫µ‡ªà‡∫° Nodes
    workflow.add_node("check_menu", check_menu)
    workflow.add_node("check_rice", check_rice)
    workflow.add_node("check_vegetables", check_vegetables) 
    workflow.add_node("check_spices", check_spices)
    workflow.add_node("cook_food", cook_food)
    workflow.add_node("deliver", deliver)
    
    # ‡∫Å‡∫≥‡∫ô‡∫ª‡∫î Edges (‡∫Å‡∫≤‡∫ô‡ªÄ‡∫ä‡∫∑‡ªà‡∫≠‡∫°‡∫ï‡ªç‡ªà)
    # Super-step 1 ‚Üí Super-step 2
    workflow.set_entry_point("check_menu")
    workflow.add_edge("check_menu", "check_rice")
    workflow.add_edge("check_menu", "check_vegetables")
    workflow.add_edge("check_menu", "check_spices")
    
    # Super-step 2 ‚Üí Super-step 3
    workflow.add_edge("check_rice", "cook_food")
    workflow.add_edge("check_vegetables", "cook_food")
    workflow.add_edge("check_spices", "cook_food")
    
    # Super-step 3 ‚Üí Super-step 4
    workflow.add_edge("cook_food", "deliver")
    workflow.add_edge("deliver", END)
    
    # Compile ‡∫î‡ªâ‡∫ß‡∫ç checkpointer
    app = workflow.compile(checkpointer=memory)
    display(Image(app.get_graph().draw_mermaid_png()))
    return app

# ‡∫Å‡∫≤‡∫ô‡∫ô‡∫≥‡ªÉ‡∫ä‡ªâ
def main():
    # ‡∫™‡ªâ‡∫≤‡∫á Graph
    app = create_order_graph()
    
    # ‡∫™‡ªâ‡∫≤‡∫á thread_id ‡∫™‡∫≥‡∫•‡∫±‡∫ö session
    thread_config = {"configurable": {"thread_id": "order_124"}}
    
    # State ‡ªÄ‡∫ö‡∫∑‡ªâ‡∫≠‡∫á‡∫ï‡∫ª‡ªâ‡∫ô
    initial_state = {
        "order": "‡∫ú‡∫±‡∫î‡ªÑ‡∫ó",
        "ingredients_checked": False,
        "food_cooked": False,
        "delivered": False,
        "status": "‡ªÄ‡∫•‡∫µ‡ªà‡∫°‡∫™‡∫±‡ªà‡∫á‡∫≠‡∫≤‡∫´‡∫≤‡∫ô",
        "rice_available": False, 
        "vegetables_available": False, 
        "spices_available": False,
        "messages": []
    }
    
    print("=" * 50)
    print("üöÄ ‡ªÄ‡∫•‡∫µ‡ªà‡∫°‡∫Å‡∫≤‡∫ô‡∫™‡∫±‡ªà‡∫á‡∫≠‡∫≤‡∫´‡∫≤‡∫ô")
    print("=" * 50)
    
    # ‡ªÄ‡∫Æ‡∫±‡∫î‡∫ß‡∫Ω‡∫Å
    result = app.invoke(initial_state, config=thread_config)
    
    print("\n" + "=" * 50)
    print("üìã ‡∫™‡∫∞‡∫ñ‡∫≤‡∫ô‡∫∞‡∫™‡∫∏‡∫î‡∫ó‡ªâ‡∫≤‡∫ç:")
    print(f"   ‡∫≠‡∫≤‡∫´‡∫≤‡∫ô: {result['order']}")
    print(f"   ‡∫™‡∫∞‡∫ñ‡∫≤‡∫ô‡∫∞: {result['status']}")
    print(f"   ‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö - ‡ªÄ‡∫Ç‡∫ª‡ªâ‡∫≤: {'‚úÖ' if result.get('rice_available') else '‚ùå'}")
    print(f"   ‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö - ‡∫ú‡∫±‡∫Å: {'‚úÖ' if result.get('vegetables_available') else '‚ùå'}")
    print(f"   ‡∫ß‡∫±‡∫î‡∫ñ‡∫∏‡∫î‡∫¥‡∫ö - ‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫õ‡∫∏‡∫á: {'‚úÖ' if result.get('spices_available') else '‚ùå'}")
    print(f"   ‡∫Å‡∫ß‡∫î‡∫™‡∫∞‡∫ï‡ªã‡∫≠‡∫Å: {result['ingredients_checked']}")
    print(f"   ‡∫õ‡∫∏‡∫á‡ªÅ‡∫•‡ªâ‡∫ß: {result['food_cooked']}")
    print(f"   ‡∫™‡∫ª‡ªà‡∫á‡ªÅ‡∫•‡ªâ‡∫ß: {result['delivered']}")
    print(f"   ‡∫Ç‡ªç‡ªâ‡∫Ñ‡∫ß‡∫≤‡∫°‡∫ó‡∫±‡∫á‡ªù‡∫ª‡∫î: {result['messages']}")
    

if __name__ == "__main__":
    main()
```

```python
api_wrapper_arxiv=ArxivAPIWrapper(top_k_results=2,doc_content_chars_max=500)
arxiv=ArxivQueryRun(api_wrapper=api_wrapper_arxiv) 
arxiv.invoke("Attention iss all you need")
```

```python
api_wrapper_wiki=WikipediaAPIWrapper(top_k_results=1,doc_content_chars_max=500)
wiki=WikipediaQueryRun(api_wrapper=api_wrapper_wiki)
wiki.invoke("What is machine learning")
```

```python
tavily = TavilySearchResults()
tavily.invoke("Provide me the recent AI news for july 27th 2025")
```

```python
### Custom Functions
def multiply(a: int, b: int) -> int:
    """Multiply a and b.

    Args:
        a: first int
        b: second int
    """
    return a * b

# This will be a tool
def add(a: int, b: int) -> int:
    """Adds a and b.

    Args:
        a: first int
        b: second int
    """
    return a + b

def divide(a: int, b: int) -> float:
    """Divide a and b.

    Args:
        a: first int
        b: second int
    """
    return a / b
```

```python
#set Tools
tools=[arxiv,wiki,tavily,add,divide,multiply]
```

```python
from langchain_groq import ChatGroq
llm=ChatGroq(model="qwen/qwen3-32b")
llm_with_tools = llm.bind_tools(tools) 
```

```python
### Node definition
def tool_calling_llm(state:State):
    return {"messages":[llm_with_tools.invoke(state["messages"])]}

# Build graph
builder = StateGraph(State)
builder.add_node("tool_calling_llm", tool_calling_llm)
builder.add_node("tools", ToolNode(tools))

builder.add_edge(START, "tool_calling_llm")
builder.add_conditional_edges(
    "tool_calling_llm",
    # If the latest message (result) from assistant is a tool call -> tools_condition routes to tools
    # If the latest message (result) from assistant is a not a tool call -> tools_condition routes to END
    tools_condition,
)
builder.add_edge("tools","tool_calling_llm") 
```

```python

memory=MemorySaver()
graph_memory = builder.compile(checkpointer=memory)

# View
display(Image(graph_memory.get_graph().draw_mermaid_png()))
```
